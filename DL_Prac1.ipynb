```python
t1 = torch.tensor(1.0, requires_grad=True)
t2 = torch.tensor(2.0, requires_grad=True)
```

```python
z= 100*t1*t2
z.backward()
```

```python
print("dz/dt1: ",t1.grad.data)
print("dz/dt2: ",t2.grad.data)
```
